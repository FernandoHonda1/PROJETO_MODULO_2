```python
class Roupa:
    
    def __init__(self, esquerda = None, direita = None, meio = None, texto = None):
        self.esquerda = esquerda
        self.direita = direita
        self.meio = meio
        self.texto = texto
    
    def pergunta_preco(self):
        
        preco = input('PREÇO: "Alto"/"Médio"/"Baixo": ')
         
        if preco.upper() == 'ALTO': 
            self.esquerda.pergunta_estacao()
        elif preco.upper() == 'MÉDIO' or preco.upper() == 'MEDIO':
            self.meio.pergunta_estacao()
        elif preco.upper() == 'BAIXO': 
            self.direita.pergunta_estacao()
        
        else:
            print('----------------')
            print('ENTRADA INVÁLIDA')
            print('----------------')
            self.pergunta_preco()
            
    def pergunta_estacao(self):
        
        estacao = input('ESTAÇÃO: "Verão"/"Inverno": ')
                
        if estacao.upper() == 'VERÃO' or estacao.upper() == 'VERAO':
            self.esquerda.pergunta_faixa_etaria()
        elif estacao.upper() == 'INVERNO':
            self.direita.pergunta_faixa_etaria()
        
        else:
            print('----------------')
            print('ENTRADA INVÁLIDA')
            print('----------------')
            self.pergunta_estacao()
        
    def pergunta_faixa_etaria(self):

        faixa = input('FAIXA ETÁRIA: "Adulto"/"Criança": ')
               
        if faixa.upper() == 'ADULTO':
            print(' ')
            print(self.esquerda.texto)
            print(' ')
            
            decisao = input('Deseja: "Continuar" / "Encerrar": ')
            if decisao.upper() == 'CONTINUAR':
                consulta()
            if decisao.upper() == 'ENCERRAR':
                print('ENCERRADO')
        
        elif faixa.upper() == 'CRIANÇA' or faixa.upper() == 'CRIANCA':
            print(' ')
            print(self.direita.texto)
            print(' ')
            
            decisao = input('Deseja: "Continuar" / "Encerrar": ')
            if decisao.upper() == 'CONTINUAR':
                consulta()
            if decisao.upper() == 'ENCERRAR':
                print('ENCERRADO')
        else:
            print('----------------')
            print('ENTRADA INVÁLIDA')
            print('----------------')
            self.pergunta_faixa_etaria()

def consulta():  
    raiz.pergunta_preco()
    

adulto6 = Roupa(texto = 'Roupas de valor baixo, para adultos, no inverno, irão vender muito!!!') 
crianca6 = Roupa(texto = 'Roupas de valor baixo, para crianças, no inverno, venderam muito')
adulto5 = Roupa(texto = 'Roupas de valor baixo, para adultos, no verão, venderam muito')
crianca5 = Roupa(texto = 'Roupas de valor baixo, para crianças, no verão, venderam muito')
adulto4 = Roupa(texto = 'Roupas de valor médio, para adultos, no inverno, irão vender muito!!!')  
crianca4 = Roupa(texto = 'Roupas de valor médio, para crianças, no inverno, venderam pouco')
adulto3 = Roupa(texto = 'Roupas de valor médio, para adultos, no verão, venderam muito')
crianca3 = Roupa(texto = 'Roupas de valor médio, para crianças, no verão, venderam pouco')
adulto2 = Roupa(texto = 'Roupas de valor alto, para adultos, no inverno, venderam muito')
crianca2 = Roupa(texto = 'Roupas de valor alto, para crianças, no inverno, venderam pouco')
adulto1 = Roupa(texto = 'Roupas de valor alto, para adultos, no verão, venderam pouco')
crianca1 = Roupa(texto = 'Roupas de valor alto, para crianças, no verão, venderam pouco')

inverno_baixo = Roupa(esquerda = adulto6, direita = crianca6)
verao_baixo = Roupa(esquerda = adulto5, direita = crianca5)
inverno_medio = Roupa(esquerda = adulto4, direita = crianca4)
verao_medio = Roupa(esquerda = adulto3, direita = crianca3)
inverno_alto = Roupa(esquerda = adulto2, direita = crianca2)
verao_alto = Roupa(esquerda = adulto1, direita = crianca1)

baixo = Roupa(esquerda = verao_baixo, direita = inverno_baixo)
medio = Roupa(esquerda = verao_medio, direita = inverno_medio)
alto = Roupa(esquerda = verao_alto, direita = inverno_alto)

raiz = Roupa(esquerda = alto, meio = medio, direita = baixo) 
```
